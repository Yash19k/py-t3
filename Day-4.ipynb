```python
import string
l=[]
a=string.ascii_uppercase
with open("demo.txt","w+") as f:
    while(True):
        str=input("Enter : ")
        f.write(str)
        if(str=="end"):
            break
        else:
            if(str[0] in a):
                l.append(str)
for i in l:
    print(i)
```

```python
f2=open("demo.txt")
f1=open("test.txt")
f3=open("final.txt","w+")
b1=True
b2=True
while(True):
    a=f1.readline()
    if(a!=""):
        f3.write(a)
    if(a=="" and b1):
        f3.write("\n")
        b1=False
    b=f2.readline()
    if(b!=""):
        f3.write(b)
    if(b=="" and b2):
        f3.write("\n")
        b2=False
    if(a=="" and b==""):
        break
f3.seek(0)
print(f3.readlines())
```
