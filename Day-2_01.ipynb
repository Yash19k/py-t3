```python
details={"name":"alice","age":"25"}
with open("demo.txt","w+") as f:
    for i in details.keys():
        f.write(i+":"+details[i]+"\n")
with open("demo.txt","r") as f:
    print(f.read())
```

```python
f=open("demo.txt","r+")
a=f.readlines()
print(a)
with open("demo.txt","w+") as f1:
    for i in a:
        t=i.split(" ")
        for j in range(len(t)):
            if (t[j]=="java"):
                t[j]="python"
        print(t)
        for i in range(len(t)):
            t[i]=t[i]+" "
        f1.writelines(t)
```
